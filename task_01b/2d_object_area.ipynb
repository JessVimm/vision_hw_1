```python
import os

folder_path = "."  # Current directory

for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        filepath = os.path.join(folder_path, filename)
        # Ones counter
        total_ones = 0

        with open(filepath, "r") as obj: # Read .txt
            for line in obj:
                total_ones += line.count("1") # Get the count of ones in each line
        # Output format
        obj_name = filename.replace("_br.txt", "")
        print(f"{obj_name}: {total_ones} number of ones.")
```
