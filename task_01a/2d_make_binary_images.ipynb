# Make binary images

```python
import os

for filename in os.listdir('.'):
    if filename.lower().endswith('.gif'):
        with open(filename, 'rb') as f:
            binary_data = f.read()
            print("Read gif as binary.")
        output_filename = filename.rsplit('.', 1)[0] + '.bin'
        with open(output_filename, 'wb') as f:
            f.write(binary_data)
            print("Saved gif as bin.")
```

# Create binary files with 0s and 1s as .txt file

```python
import os

for filename in os.listdir('.'):
    if filename.lower().endswith('.gif'):
        with open(filename, 'rb') as f:
            binary_data = f.read()
        bits = ''.join(f'{byte:08b}' for byte in binary_data)
        txt_filename = filename.rsplit('.', 1)[0] + '.txt'
        with open(txt_filename, 'w') as f:
            f.write(bits)
```

# >Create files of 0s and 1s in .txt file with dimensions exactily as in the image

```python
from PIL import Image
import os

for filename in os.listdir('.'):
    if filename.lower().endswith('.gif'):
        with Image.open(filename) as img:
            img = img.convert('L')  # Convert to grayscale
            width, height = img.size
            pixels = img.load()
            txt_filename = filename.rsplit('.', 1)[0] + '_br.txt'
            with open(txt_filename, 'w') as f:
                for y in range(height):
                    row_bits = ''.join(f'{pixels[x, y]:08b}' for x in range(width))
                    f.write(row_bits + '\n')
```
